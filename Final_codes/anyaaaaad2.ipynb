```python
def metrics(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f'Mean Squared Error: {mse}')
    print(f'R^2 Score: {r2}')
    print(f'Mean Absolute Error: {mae}')
```

```python
import re

def extract_engine_att(engine_str):
    horsepower = re.search(r'(\d+\.\d+)HP|\d+\.\d+', engine_str)
    displacement = re.search(r'(\d+\.\d+L|\d+\.\d+ Liter)', engine_str)
    return horsepower.group(1) if horsepower else '',\
           displacement.group(1) if displacement else ''


df[['Horsepower', 'Engine_Displacement']] = df['engine'].apply(extract_engine_att).apply(pd.Series)
df=df.drop(['engine'], axis=1)
```

```python
def extract_base_colors(df):
    
    base_colors = [
        'black', 'white', 'gray', 'silver', 'brown', 'red', 'blue', 'green',
         'orange', 'yellow']

    df['int_col'] = df['int_col'].str.lower()
    df['ext_col'] = df['ext_col'].str.lower()
    
    def find_base_color(text):
        for color in base_colors:
            if color in text:
                return color
        return 'other'
    
    df['int_col'] = df['int_col'].apply(find_base_color)
    df['ext_col'] = df['ext_col'].apply(find_base_color)

    return df


df = extract_base_colors(df)


df[['int_col', 'ext_col']].head()
```

CatBoost

```python
cat_features = [6, 8, 10, 11]


# Hiperparaméterek hálója
param_grid = {
    'iterations': [100, 300, 500],  # Boosting iterációk száma
    'learning_rate': [0.01, 0.03, 0.1],  # Tanulási ráta
    'depth': [4, 6, 8, 10],  # Fa mélysége
    'l2_leaf_reg': [1, 3, 5, 7],  # L2 regulárizáció erőssége
}

# CatBoostRegressor alapértelmezett modell
cat_model = CatBoostRegressor(random_state=42, verbose=0)

# GridSearchCV beállítása
grid_search = GridSearchCV(
    estimator=cat_model,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',  # R² pontszám alapján történő kiértékelés
    cv=5,  # 5-szörös kereszt-validáció
    n_jobs=-1,  # Több processzor használata a gyorsítás érdekében
    verbose=1  # GridSearch kimenetének engedélyezése
)

# Futtatás
grid_search.fit(X_train, y_train, cat_features=cat_features)

# Legjobb paraméterek és eredmények
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
```

```python
cat_features = [7, 9, 10, 15]


best_params = {
    'depth': 8,
    'iterations': 500,
    'l2_leaf_reg': 5,
    'learning_rate': 0.1
}


best_cat = CatBoostRegressor(**best_params, feature_weights={'accident': 5}, random_state=42, verbose=0)

X_train_part, X_valid, y_train_part, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
eval_set2=[(X_valid, y_valid)]

best_cat.fit(X_train, y_train, cat_features=cat_features, eval_set=eval_set2, verbose=True)

y_pred = best_cat.predict(X_test)

train_rmse = mean_squared_error(y_train, best_cat.predict(X_train), squared=False)
test_rmse = mean_squared_error(y_test, best_cat.predict(X_test), squared=False)
train_r2 = r2_score(y_train, best_cat.predict(X_train))
test_r2 = r2_score(y_test, best_cat.predict(X_test))

print(f"Train RMSE: {train_rmse}")
print(f"Test RMSE: {test_rmse}")
print(f"Train R2: {train_r2}")
print(f"Test R2: {test_r2}")
```
